## 1. Count Word Frequency

**Question:** Write a Python function that takes a sentence and returns a dictionary showing how many times each word appears.

Example:

```python
sentence = "python is easy and python is powerful"
```

Expected output:

```python
{'python': 2, 'is': 2, 'easy': 1, 'and': 1, 'powerful': 1}
```

````python
# Starter code

sentence = "python is easy and python is powerful"

def count_words(sentence):
    # write your code here
    pass

print(count_words(sentence))
````

````python
# Solution

sentence = "python is easy and python is powerful"

def count_words(sentence):
    words = sentence.split()
    result = {}

    for word in words:
        if word in result:
            result[word] += 1
        else:
            result[word] = 1

    return result

print(count_words(sentence))
````

## 7. Reverse a List Without `reverse()` or `[::-1]`

**Question:** Reverse a list using a loop.

Example:

```python
numbers = [10, 25, 3, 99, 44, 18]
```

Expected output:

```python
[18, 44, 99, 3, 25, 10]
```

````python
# Starter code

numbers = [10, 25, 3, 99, 44, 18]

reversed_list = []

# write your loop here

print(reversed_list)
````

````python
# Solution

numbers = [10, 25, 3, 99, 44, 18]

reversed_list = []

for i in range(len(numbers) - 1, -1, -1):
    reversed_list.append(numbers[i])

print(reversed_list)
````

### Understanding `range(len(numbers) - 1, -1, -1)`

For this list:

```python
numbers = [10, 25, 3, 99, 44, 18]
```

`len(numbers)` is `6`.

Last index is:

```python
len(numbers) - 1 = 5
```

So:

```python
range(5, -1, -1)
```

produces:

```python
5, 4, 3, 2, 1, 0
```

It includes index `5`, so it does not miss the last value.

## 9. Find Second Largest Number Without `sort()` or `max()`

**Question:** Find the second largest number in a list.

Example:

```python
numbers = [10, 25, 3, 99, 44, 18, 7]
```

Expected output:

```python
44
```

````python
# Starter code

numbers = [10, 25, 3, 99, 44, 18, 7]

largest = numbers[0]
second_largest = numbers[0]

# write your logic here

print(second_largest)
````

````python
# Solution

numbers = [10, 25, 3, 99, 44, 18, 7]

largest = numbers[0]
second_largest = numbers[0]

for num in numbers:
    if num > largest:
        second_largest = largest
        largest = num
    elif num > second_largest and num < largest:
        second_largest = num

print(second_largest)
````

## 10. Check Palindrome Without `[::-1]`

**Question:** Check whether a word is a palindrome.

A palindrome reads the same forward and backward.

Example:

```python
word = "madam"
```

Expected output:

```python
Palindrome
```

````python
# Starter code

word = "madam"

is_palindrome = True

# write your loop here

if is_palindrome:
    print("Palindrome")
else:
    print("Not Palindrome")
````

````python
# Solution

word = "madam"

is_palindrome = True

for i in range(0, len(word) // 2):
    if word[i] != word[len(word) - 1 - i]:
        is_palindrome = False
        break

if is_palindrome:
    print("Palindrome")
else:
    print("Not Palindrome")
````

## 12. Count Consonants in a String

**Question:** Count how many consonants are in a string.

Example:

```python
text = "machine learning"
```

Expected output:

```python
9
```

Important: Do not count spaces, numbers, or symbols.

````python
# Starter code

text = "machine learning"

vowels = "aeiou"
count = 0

# write your loop here

print(count)
````

````python
# Solution

text = "machine learning"

vowels = "aeiou"
count = 0

for char in text:
    if char.isalpha() and char not in vowels:
        count += 1

print(count)
````

### Understanding `char.isalpha() and char not in vowels`

```python
char.isalpha()
```

checks whether the character is a letter.

Examples:

```python
"a".isalpha()  # True
"m".isalpha()  # True
" ".isalpha()  # False
"1".isalpha()  # False
```

```python
char not in vowels
```

checks whether the character is not a vowel.

So this condition:

```python
if char.isalpha() and char not in vowels:
```

means:

**Count only alphabet letters that are not vowels.**
